```python
%matplotlib inline
import matplotlib.pyplot as plt
```

#### Generate Random Data

```python
import numpy as np
import pandas as pd

COL_NUM = 10

# initial column
df = pd.DataFrame(np.random.random((100, 1)), columns=['col_a'])

# add more columns
for i in range(COL_NUM - 1):
    col_name = f"col_{chr(ord('a') + i + 1)}"
    df[col_name] = df['col_a'] ** i + np.random.random() * (0.9 ** i)
    
df.head(10)
```

#### Plot the results

```python
plt.figure(figsize=(12, 5))
for i in range(COL_NUM):
    plt.hist(df[f"col_{chr(ord('a') + i)}"], bins=7, rwidth=0.9, alpha=0.3)
plt.legend([f"col_{chr(ord('a') + i)}" for i in range(COL_NUM)])
```
